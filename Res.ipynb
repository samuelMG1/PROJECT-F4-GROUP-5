```python
import json
import pandas as pd

# List to store parsed JSON data
json_data = []
lines_to_read = 54380  # Adjust the number of lines you want to read

# Open and read the JSON file, limiting the number of lines
file_path = r"C:\Users\Administrator\Downloads\yelp_dataset\yelp_academic_dataset_review.json"

# Read the file and extract the specified number of lines
with open(file_path, 'r', encoding='utf-8') as file:
    for i, line in enumerate(file):
        if i >= lines_to_read:  # Stop after reading the specified number of lines
            break
        json_object = json.loads(line)
        json_data.append(json_object)

# Convert the extracted JSON data into a pandas DataFrame
review_df = pd.DataFrame(json_data)

# Save the DataFrame to a CSV file for easier future use
output_csv_path = r'C:\Users\Administrator\Downloads\yelp_dataset\review_sample.csv'
review_df.to_csv(output_csv_path, index=False)

print("Sample data saved to:", output_csv_path)
```
