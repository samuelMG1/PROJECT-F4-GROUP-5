## FIRST DATASET CONTAINING BUSINESS INFORMATION

```python
# File Path to JSON file
file_path = r"C:\Users\Administrator\Downloads\yelp_dataset\yelp_academic_dataset_business.json"

# Load the JSON file into a DataFrame,
business = pd.read_json(file_path, lines=True)

# Displaying first 5 entries
business.head()
```

```python
business.info()
```

```python
# Filter the DataFrame to keep only rows that contain the word "restaurant"
restaurants_df = business[business['categories'].str.contains("restaurant", case=False, na=False)]

# Display the first few rows of the filtered DataFrame
restaurants_df.head()
```

```python
restaurants_df.info()
```

```python
# Saving restaurant data to CSV
restaurants_df.to_csv(r"C:\Users\Administrator\Downloads\yelp_dataset\restaurants.csv", index= False)
```

## SECOND DATASET CONTAINING REVIEWS INFORMATION

```python
# File Path to  JSON file
file_path = r"C:\Users\Administrator\Downloads\yelp_dataset\yelp_academic_dataset_review.json"

# Define a chunk size
chunk_size = 10000

# Initialize an empty list to hold chunks
chunks = []

# Read the JSON file in chunks
with pd.read_json(file_path, lines=True, chunksize=chunk_size) as reader:
    for chunk in reader:
        # Process each chunk (e.g., filter, aggregate)
        chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
reviews = pd.concat(chunks, ignore_index=True)
```

```python
# Information of the reviews dataset
reviews.info()
```

```python
# Getting the unique values in the restaurant_df's business Id column
restaurants_ids = restaurants_df['business_id'].unique()
len(restaurants_ids)
```

```python
# Filtering the reveiws dataframe to include only reviews related to restaurants
rest_reviews = reviews[reviews['business_id'].isin(restaurants_ids)]
```

```python
# Getting resturant reviews infor
rest_reviews.info()
```

```python
 # first 5 entries in restaurant reviews
rest_reviews.head()
```

```python
# Timeframe of the dataset
rest_reviews.date.min(), rest_reviews.date.max()
```

```python
# Starting date of the dataframe to get 5 year worth of data
start_date = '2017-01-01 00:00:00'

# Getting 5 years worth of reviews
five_year_reviews = rest_reviews[(rest_reviews['date'] >= start_date)]

# first five entries
five_year_reviews.head()
```

```python
# Getting information on the 5 year reviews dataset
five_year_reviews.info()
```

```python
# Filtering to have only the relevant dataset
five_year_reviews=five_year_reviews[['user_id','business_id','stars','date']]

five_year_reviews.info()
```

```python
# saving the one_year_reviews to users.csv
five_year_reviews.to_csv(r"C:\Users\Administrator\Downloads\yelp_dataset\users.csv", index= False)
```
